We have time series data for CPU, memory, and disk usage attached in a CSV-encoded text file. We want to analyze the data to better understand what is happening so we can inform our users. Write a program that will read in the full data, bucket each series into five minute intervals (aligned to the hour), and calculate the minimum, maximum, mean, and standard deviation of the data points in each bucket. Your program should print out the calculated statistics for each bucket. Hint: Not all of the buckets will have data.

```python
csv_file = r'C:\usf\Summer works\obama\timeseries.csv'
```

```python
output_file=r'C:\usf\Summer works\obama\statistics.csv'
```

```python
import pandas as pd
df = pd.read_csv(csv_file)
```

```python
df
```

```python
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)
```

```python
df_cpu=df[df['label'] == 'cpu']
df_disk=df[df['label'] == 'disk']
df_memory=df[df['label'] == 'memory']
```

```python
cpu_std=df_cpu.groupby(pd.TimeGrouper('5Min'))['value'].std().reset_index().rename(columns={'value':'cpu_std'})
cpu_mean=df_cpu.groupby(pd.TimeGrouper('5Min'))['value'].mean().reset_index().rename(columns={'value':'cpu_mean'})
cpu_min=df_cpu.groupby(pd.TimeGrouper('5Min'))['value'].min().reset_index().rename(columns={'value':'cpu_min'})
cpu_max=df_cpu.groupby(pd.TimeGrouper('5Min'))['value'].max().reset_index().rename(columns={'value':'cpu_max'})
disk_std=df_disk.groupby(pd.TimeGrouper('5Min'))['value'].std().reset_index().rename(columns={'value':'disk_std'})
disk_mean=df_disk.groupby(pd.TimeGrouper('5Min'))['value'].mean().reset_index().rename(columns={'value':'disk_mean'})
disk_min=df_disk.groupby(pd.TimeGrouper('5Min'))['value'].min().reset_index().rename(columns={'value':'disk_min'})
disk_max=df_disk.groupby(pd.TimeGrouper('5Min'))['value'].max().reset_index().rename(columns={'value':'disk_max'})
memory_std=df_memory.groupby(pd.TimeGrouper('5Min'))['value'].std().reset_index().rename(columns={'value':'memory_std'})
memory_mean=df_memory.groupby(pd.TimeGrouper('5Min'))['value'].mean().reset_index().rename(columns={'value':'memory_mean'})
memory_min=df_memory.groupby(pd.TimeGrouper('5Min'))['value'].min().reset_index().rename(columns={'value':'memory_min'})
memory_max=df_memory.groupby(pd.TimeGrouper('5Min'))['value'].max().reset_index().rename(columns={'value':'memory_max'})
```

```python
result=pd.concat([cpu_mean.set_index('time'),cpu_max.set_index('time'),cpu_min.set_index('time'),cpu_std.set_index('time'),memory_mean.set_index('time'),memory_max.set_index('time'),memory_min.set_index('time'),memory_std.set_index('time'),disk_mean.set_index('time'),disk_max.set_index('time'),disk_min.set_index('time'),disk_std.set_index('time')], axis=1, join='inner').reset_index()
```

```python
result.to_csv(output_file)
```
